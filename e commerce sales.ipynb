```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default="plotly_white"
```

```python
data=pd.read_csv("Sample - Superstore.csv",encoding='latin-1')
```

```python
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
```

```python
data.info()
```

```python
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Month']=data['Order Date'].dt.month
```

```python
data.head()
```

```python
data['Order Year']=data['Order Date'].dt.year
data['Order day']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
sales_by_month=data.groupby('Order Month') ['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig=px.line(sales_by_month,
           x='Order Month',
           y='Sales',
           title='monthly sales')

fig.show()
```

#  sales by category

```python
sales_by_category=data.groupby('Category') ['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig=px.pie(sales_by_category,
           values='Sales',
           names='Category',
           
           title='category by sales')

fig.show()
           
```

```python
sales_by_subcategory=data.groupby('Sub-Category') ['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
fig=px.bar(sales_by_subcategory,
           x='Sub-Category',
           y='Sales',
           title='sales by subcategory')

fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show()
```

```python
profit_by_month=data.groupby('Order Month')['Profit'].sum().reset_index()
```

```python
profit_by_month
```

```python
fig=px.line(profit_by_month,
            y='Profit',
            x='Order Month',
            title='profit by month')

fig.show()
```

```python
category_by_profit=data.groupby('Category')['Profit'].sum().reset_index()
```

```python
category_by_profit
```

```python
fig=px.pie(category_by_profit,
           values='Profit',
           names='Category',
           title='profit by category')

fig.show()
```

```python
subcategory_by_profit=data.groupby('Sub-Category')['Profit'].sum().reset_index()
```

```python
subcategory_by_profit
```

```python
fig=px.line(subcategory_by_profit,
           y='Profit',
           x='Sub-Category',
           title='sub profit by category')

fig.show()
```

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales' :'sum','Profit':'sum'}).reset_index()
```

```python
sales_profit_by_segment
```

```python


fig=go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='sales'))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit'))

fig.update_layout(title='sales profit by customer segment',
                  xaxis_title='customer_segment',yaxis_title='amount')


fig.show()
```

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales' :'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['sales_profit_ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','sales_profit_ratio']])
```
